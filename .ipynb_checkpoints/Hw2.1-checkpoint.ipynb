```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tbl = pd.read_html("http://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C", match="Коефіцієнт народжуваності в регіонах України", thousands=".", decimal=",")[0]
```

```python
print(tbl.head())
```

```python
tbl.shape
```

```python
tbl.replace('—', np.nan, inplace=True)
```

```python
tbl.dtypes
```

```python
tbl[["2014", "2019"]] = tbl[["2014", "2019"]].apply(pd.to_numeric)
tbl.dtypes
```

```python
tbl.isnull().sum()
```

```python
tbl
```

```python
tbl.drop([27])
```

```python
tbl = tbl.fillna(tbl.loc[:,"1950" :].mean())
tbl
```

```python
tbl[tbl["2019"] > tbl["2019"][27]][["Регіон", "2019"]]
```

```python
tbl.loc[tbl["2014"].idxmax()]
```

```python
import matplotlib.pyplot as plt
born_rate_2019 = pd.Series(list(tbl["2019"]), index=list(tbl["Регіон"]))

born_rate_2019.plot(kind="bar", color=['b', 'y'])
plt.grid()
plt.show()
```

```python
sns.set_style("whitegrid")
```

```python
plt.plot(tbl.columns[1:], list(tbl.iloc[27][1:]))
plt.xlabel('Year', fontsize='small', color='midnightblue')
plt.ylabel('Born rate', fontsize='small', color='midnightblue')
plt.title('Born rate in Ukraine', fontsize=15)
plt.show()
```

```python
labels = tbl["Регіон"][:27]
data = tbl["2019"][:27]

plt.pie(
    data,
    labels=labels,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.3,
)
plt.title("Born rate per regions in Ukraine")

plt.show()
```
